##1. Data Cleaning
As per initial observation from data, looks there are multiple NULL values.  Try to plot the Null value % to understand further.
Try to check and remove columns where NULL / NA value percentage is more than 30%

````python

def find_na_value_percentage(df: pd.DataFrame):
    if loan_data.isnull().sum().sum() != 0:
      percentage_value = (loan_data.isnull().sum() / len(df)) * 100
      na_value_data = (loan_data.isnull().sum() / len(df)) * 100
      na_value_data = na_value_data.drop(na_value_data[na_value_data == 0].index).sort_values(ascending=False)
      missing_value_data = pd.DataFrame({'column_name': df.columns,
                                 'percent_missing': percentage_value})

      missing_data_condition = missing_value_data[(missing_value_data['percent_missing']<=30.00) ].index
      missing_value_data = missing_value_data.drop(missing_data_condition)
      return missing_value_data
missing_data = find_na_value_percentage(loan_data)
na_list = missing_data['column_name'].tolist()
````

##1.1.Standardize parameters
Below parameters needs to be converted and standardized as these are used further down in analysis
- emp_length
- int_rate
- term

````python
#emp_length has many values as 0.
#Firstly fill all NA values with 0
loan_data['emp_length'].fillna('0', inplace = True)
# As per data present the value ranges from <1 to 10+
# for values <1 assuming it to be 0. and values >10 or 10+ assuming it to be 10
# Additionally discard 'years' and convert this field to numberical value
#extract integer value
loan_data['emp_length'] = loan_data['emp_length'].apply(lambda x: 0 if "<" in x else (x.split('+')[0] if "+" in x else x.split()[0]))
loan_data['emp_length'] = pd.to_numeric(loan_data['emp_length'])
````

````python
#Remove some unwanted characters like % etc. on numerical values
loan_data['int_rate'] = loan_data['int_rate'].str.rstrip('%')

#COnvert DataType for further analysis
loan_data['int_rate'] = loan_data['int_rate'].astype('float64')
````

````python
#term is object as it has some text associated like 36 Months, 60 Months
#Convert this to numerical data
def convert_term(var):
  var = var.rstrip(' months')
  var = int(var)
  return var
loan_data['term'] = loan_data['term'].apply(convert_term)
````
